## Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python

# Load the datasets
spx = pd.read_csv('yahoo_SPX.csv')
ndx = pd.read_csv('yahoo_NDX.csv')
btc = pd.read_csv('yahoo_BTC.csv')
eth = pd.read_csv('yahoo_ETH.csv')
bnb = pd.read_csv('yahoo_BNB.csv')
dow = pd.read_csv('yahoo_DJIA.csv')
```

```python
# Function to preprocess each dataset
def preprocess_data(df, date_col, drop_cols, rename_cols):
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.drop(columns=drop_cols)
    df = df.sort_values(by=date_col).reset_index(drop=True)
    df.rename(columns=rename_cols, inplace=True)
    return df

# Preprocess each dataset
btc = preprocess_data(btc, 'Date', ['Open', 'High', 'Low', 'Volume', 'Adj Close'], {'Close': 'Close_BTC'})
eth = preprocess_data(eth, 'Date', ['Open', 'High', 'Low', 'Volume', 'Adj Close'], {'Close': 'Close_ETH'})
bnb = preprocess_data(bnb, 'Date', ['Open', 'High', 'Low', 'Volume', 'Adj Close'], {'Close': 'Close_BNB'})
dow = preprocess_data(dow, 'Date', ['Open', 'High', 'Low', 'Volume', 'Adj Close'], {'Close': 'Close_DOW'})
ndx = preprocess_data(ndx, 'Date', ['Open', 'High', 'Low', 'Volume', 'Adj Close'], {'Close': 'Close_NDX'})
spx = preprocess_data(spx, 'Date', ['Open', 'High', 'Low', 'Volume', 'Adj Close'], {'Close': 'Close_SPX'})

# Merging datasets on Date
merged_data = btc.merge(eth, on='Date', suffixes=('_BTC', '_ETH'))
merged_data = merged_data.merge(bnb, on='Date', suffixes=('', '_BNB'))
merged_data = merged_data.merge(dow, on='Date', suffixes=('', '_DOW'))
merged_data = merged_data.merge(ndx, on='Date', suffixes=('', '_NDX'))
merged_data = merged_data.merge(spx, on='Date', suffixes=('', '_SPX'))

# Normalize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(merged_data.drop(columns=['Date']))
scaled_data = pd.DataFrame(scaled_data, columns=merged_data.columns.drop('Date'))
scaled_data['Date'] = merged_data['Date']
```

# Correlation and Causality

```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
#read the datasets
bnb = pd.read_csv("yahoo_BNB.csv")
btc = pd.read_csv("yahoo_BTC.csv")
eth = pd.read_csv("yahoo_ETH.csv")
ndx = pd.read_csv("yahoo_NDX.csv")
spx = pd.read_csv("yahoo_SPX.csv")
dow = pd.read_csv("yahoo_DJIA.csv")

#convert the 'Datetime' column to datetime format
bnb['Date'] = pd.to_datetime(bnb['Date'])
btc['Date'] = pd.to_datetime(btc['Date'])
eth['Date'] = pd.to_datetime(eth['Date'])
ndx['Date'] = pd.to_datetime(ndx['Date'])
spx['Date'] = pd.to_datetime(spx['Date'])
dow['Date'] = pd.to_datetime(dow['Date'])

#set 'Date' as the index
bnb.set_index('Date', inplace=True)
btc.set_index('Date', inplace=True)
eth.set_index('Date', inplace=True)
ndx.set_index('Date', inplace=True)
spx.set_index('Date', inplace=True)
dow.set_index('Date', inplace=True)
```

```python
bnb.isnull().sum()
```

### Pearson correlation

```python
# Use the common datetime index that spans all datasets to make them aligned
common_index = pd.date_range(start=max(btc.index.min(), bnb.index.min(), eth.index.min(), ndx.index.min(), spx.index.min(),  dow.index.min()), 
                             end=min(btc.index.max(), bnb.index.max(), eth.index.max(), ndx.index.max(), spx.index.max(),  dow.index.max()), 
                             freq='min')

# Reindex all dataframes to the common index and select the 'Close' column
btc = btc['Close'].reindex(common_index).fillna(method='bfill')
bnb = bnb['Close'].reindex(common_index).fillna(method='bfill')
eth = eth['Close'].reindex(common_index).fillna(method='bfill')
ndx = ndx['Close'].reindex(common_index).fillna(method='bfill')
spx = spx['Close'].reindex(common_index).fillna(method='bfill')
dow = dow['Close'].reindex(common_index).fillna(method='bfill')

# Combine the datasets into a single DataFrame
combined_data = pd.DataFrame({'NDX': ndx, 'SPX': spx, 'BTC': btc, 'BNB': bnb, 'ETH': eth, 'DOW': dow})

# Drop any remaining NaN values (if any)
combined_data = combined_data.dropna()

# Calculate the Pearson correlation
correlation = combined_data.corr()
correlation

```

### Granger Causality

```python
# Granger Causality Test
print("\n DOW-BTC:")
dow_btc = grangercausalitytests(combined_data[['DOW', 'BTC']], maxlag=1)
print("\n DOW-BNB:")
dow_bnb = grangercausalitytests(combined_data[['DOW', 'BNB']], maxlag=1)
print("\n DOW-ETH:")
dow_eth = grangercausalitytests(combined_data[['DOW', 'ETH']], maxlag=1)
print("\n NDX-BTC:")
ndx_btc = grangercausalitytests(combined_data[['NDX', 'BTC']], maxlag=1)
print("\n NDX-BNB:")
ndx_bnb = grangercausalitytests(combined_data[['NDX', 'BNB']], maxlag=1)
print("\n NDX-ETH:")
ndx_eth = grangercausalitytests(combined_data[['NDX', 'ETH']], maxlag=1)
print("\n SPX-BTC:")
spx_btc = grangercausalitytests(combined_data[['SPX', 'BTC']], maxlag=1)
print("\n SPX-BNB:")
spx_bnb = grangercausalitytests(combined_data[['SPX', 'BNB']], maxlag=1)
print("\n SPX-ETH:")
spx_eth = grangercausalitytests(combined_data[['SPX', 'ETH']], maxlag=1)
```
